### Unzipping the Dataset

```python
import zipfile
import os

zip_path = '/content/titanicDataset.zip'
extract_dir = '/content/titanic_data'

# Create the directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"Files extracted to: {extract_dir}")
print("Contents of the extracted directory:")
print(os.listdir(extract_dir))
```

### Loading and Initial Data Exploration

```python
import pandas as pd

# The extracted dataset is named 'Titanic-Dataset.csv'
dataset_path = os.path.join(extract_dir, 'Titanic-Dataset.csv')

# Load the dataset
train_df = pd.read_csv(dataset_path)

print("First 5 rows of the training data:")
display(train_df.head())

print("\nInformation about the training data (data types and non-null counts):")
train_df.info()
```

### Checking for Missing Values

```python
print("Missing values per column:")
missing_values = train_df.isnull().sum()
missing_values_percent = (train_df.isnull().sum() / len(train_df)) * 100

missing_info = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_values_percent
})

display(missing_info[missing_info['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False))
```

### Handling Missing Values

```python
# 1. Drop the 'Cabin' column due to a high percentage of missing values
train_df = train_df.drop('Cabin', axis=1)
print("Dropped 'Cabin' column.")

# 2. Impute missing 'Age' values with the median
median_age = train_df['Age'].median()
train_df['Age'].fillna(median_age, inplace=True)
print(f"Imputed missing 'Age' values with the median: {median_age:.2f}")

# 3. Impute missing 'Embarked' values with the mode
mode_embarked = train_df['Embarked'].mode()[0]
train_df['Embarked'].fillna(mode_embarked, inplace=True)
print(f"Imputed missing 'Embarked' values with the mode: {mode_embarked}")

print("\nFirst 5 rows after handling missing values:")
display(train_df.head())

print("\nRe-checking missing values after imputation:")
display(train_df.isnull().sum()[train_df.isnull().sum() > 0])
```

### Encoding Categorical Features

```python
# 1. Encode 'Sex' column: convert 'male' to 0 and 'female' to 1
train_df['Sex'] = train_df['Sex'].map({'male': 0, 'female': 1})
print("Encoded 'Sex' column to numerical values.")

# 2. One-hot encode 'Embarked' column
train_df = pd.get_dummies(train_df, columns=['Embarked'], drop_first=True)
print("One-hot encoded 'Embarked' column.")

print("\nFirst 5 rows after encoding categorical features:")
display(train_df.head())

print("\nInformation about the DataFrame after encoding:")
train_df.info()
```
